```python
#import mapping data frame
mapping=pd.read_csv(r'C:\Users\HARUN\Downloads\mapping.csv',encoding='ISO-8859-1',sep=',')
```

```python
print(mapping)
```

```python
#code for replace 0 to na in catagory_list
mapping['category_list']=mapping['category_list'].str.replace('0','na')
```

```python
#code to split the primary_sector
master_frame['primary_sector']=master_frame.loc[:,'category_list'].str.split('|').str.get(0)
master_frame
```

```python
#code for remove null rows in category_list
mapping=mapping[~pd.isnull(mapping['category_list'])]
```

```python
print(round(100*(mapping.isnull().sum()/len(mapping.index)),2))
```

```python
#code to transpote the columns
mapping=pd.melt(mapping,id_vars=['category_list'],var_name='main_sector')
mapping
```

```python
#code to fetch only true rows
mapping=mapping[mapping['value']==True]
mapping
```

```python
#code to merge master_frame,mapping data frames
master_mapping_frame1=pd.merge(master_frame,mapping,how='left',left_on='primary_sector',right_on='category_list')
master_mapping_frame1
```

```python
#code to create D1 dataframe for country code USA
D1= master_mapping_frame1.loc[(master_mapping_frame1.funding_round_type=='venture')&(master_mapping_frame1.raised_amount_usd>5000000)&(master_mapping_frame1.raised_amount_usd<15000000)&(master_mapping_frame1.country_code=='USA')]
D1
```

```python
#code for create two columns no.of investments and total amount_invested
D1['no.of_investments']=D1.groupby(["main_sector"]).main_sector.transform('count')
D1['total_amount_invested']=D1.groupby(["main_sector"]).raised_amount_usd.transform('sum')

D1.columns
```

```python
#total number of investments(count) and total amount of investment(usd)in D1 data frame
print("total no of investments USA:",D1.raised_amount_usd.count())
print("total amount of investments USA:",D1.raised_amount_usd.sum())
```

```python
#top sector based on count of investments,first,second, and third with counts
D1_investmentscount=D1.groupby(["main_sector"]).main_sector.count()

D1_investmentscount = D1_investmentscount.sort_values(ascending=False)
D1_investmentscount
```
